# Indexing & Selecting Data from pandas data Structures

```python
import pandas as pd
import numpy as np
```

```python
dates = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.randn(8, 4), index=dates, columns=['A', 'B', 'C', 'D'])
df
```

># Using basic [] for Dataframe

```python
#moving data from A to B
df[['B','A']] = df[['A','B']]
df
```

># Using .loc method to select by labeling

```python
df.loc[:, ['A', 'B']] = df[['B', 'A']].to_numpy()
df
```

```python
df['Z'] = list(range(len(df.index)))
df
```

> # Way of accesing by index label to a dataframe using iloc

```python
x = pd.DataFrame({'x': [1, 2, 3], 'y': [3, 4, 5]})
x.iloc[1] = {'x': 9, 'y': 99}
x
```

```python
x["z"] = [6,7,8]
x
```

> # Slicing ranges using []

```python
df[:5]
```

```python
df[::2]
```

```python
df[:3:-1]
```

> # Pandas purely label based indexing

```python
df1 = pd.DataFrame(np.random.randn(6, 4),index=list('abcdef'),columns=list('ABCD'))
df1
```

```python
df1.loc[['a', 'b', 'd'], :]
```

```python
df1.loc['d':, 'A':'C']
```
